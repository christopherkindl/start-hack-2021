```python
#create function to add time dimensions

def create_features(df, label=None):
    """
    Input initial df.
    Create several time dimensions, such as quarter or day of week.
    Return df with new time dimensions.
    """
    df = df.copy()
    #df['id'] = df.index
    df['date_only'] = df['date'].dt.date
    df['hour'] = df['date'].dt.hour
    df['day_of_week'] = df['date'].dt.dayofweek
    df['day_name'] = df['date'].dt.day_name()
    df['quarter'] = df['date'].dt.quarter
    df['month'] = df['date'].dt.month
    df['year'] = df['date'].dt.year
    df['day_of_year'] = df['date'].dt.dayofyear
    df['day_of_month'] = df['date'].dt.day
    df['week_of_year'] = df['date'].dt.weekofyear
    
    return df
```

## Time Series Models

**ARIMA Model**

**1) Stationarity-Test**

```python
# only keep date and one variable series
df = df[['date', 'occupancy_rate']]

# apply stationarity test and add attributes
result = adfuller(df['occupancy_rate'])
dict(zip(['adf', 'pvalue', 'usedlag', 'nobs', 'critical' 'values', 'icbest'],result))
```

We failed to reject the Null Hypothesis meaning that the series has a unit root thus is not stationary.

**2) Transform Non-Stationary to Stationary using Differencing**

Transform data to stationary so we will have an estimate for $d$ and $D$ parameters we will use in the model.

```python
df1 = df.groupby('date_only')['occupancy_rate'].mean()

# number of days occupancy rate is below 30%
print(df1[df1 < 30].count())

# number of days occupancy rate is below 20%
print(df1[df1 < 20].count())

# number of days occupancy rate is below 10%
print(df1[df1 < 10].count())
```

```python
# plot time series before transformation
y = df.groupby('date_only')['occupancy_rate'].mean()
x = df.groupby('date_only')['occupancy_rate'].mean().index # date column

fig = plt.figure(figsize=(20,10));
ax = fig.add_subplot(111);

plt.plot(x, y, marker=".", markersize=20);
plt.title('Occupancy rate per date', fontsize = 16, weight = 'bold');
plt.xlabel('Date', fontsize = 16, weight = 'bold');
plt.ylabel('Occupancy rate', fontsize = 16, weight = 'bold');
plt.show();
```

```python
df['1_difference'] = df['occupancy_rate'] - df['occupancy_rate'].shift(1)
result = adfuller(df['1_difference'].dropna())
dict(zip(['adf', 'pvalue', 'usedlag', 'nobs', 'critical' 'values', 'icbest'],result))
```

We can reject the null hypothesis.

**Model performance evaluation**

```python
def mean_absolute_percentage_error(y_true, y_pred): 
    """Calculates MAPE given y_true and y_pred"""
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

mean_absolute_percentage_error(y_true=pjme_test['occupancy_rate'], y_pred=pjme_test_fcst['yhat'])
```
